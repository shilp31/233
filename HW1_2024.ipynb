## Bayes Rules

Answer\\

Assume the notations :
\begin{align*}
    \text{Test = Positive as }T\\
    \text{Test = Negative as }\neg T\\
    \text{Disease = Positive as }D\\
    \text{Disease = Negative as }\neg D
\end{align*}
From the given data, we have 
\begin{align*}
    P(T|D) = 0.992\\
    \implies P(\neg T| D) = 0.008 \\
    P(\neg T| \neg D) = 0.992\\
    \implies P(T| \neg D) = 0.008 \\
    P(D) = 0.001
\end{align*}
From Baye's rule, we know that
\begin{align*}
    P(D|T) = \frac{P(T|D) P(D)}{P(T)}
\end{align*}
We can find $P(T)$ by applying the law of total probability.
\begin{equation*}
\begin{split}
P(T) &= P(T|D)P(D) + P(T|\neg D) P(\neg D)\\
&= 0.992*0.001 + 0.008*0.999\\
&= 0.008984
\end{split}
\end{equation*}
Substituting, we get
\begin{equation*}
P(D| T) = \frac{0.992*0.001}{0.008984} = 0.1104
\end{equation*}

Therefore the probability of having the disease given the test result is positive is $0.1104$.

After noticing the error,
\begin{align*}
    P(T| D) = 0.92\\
    P(T) = 0.92*0.001 + (1-0.92)*0.999 = 0.08084
\end{align*}
\begin{equation*}
P(D| T) = \frac{0.92*0.001}{0.08084} = 0.01138
\end{equation*}
Therefore, the probability of having the disease is reduced from $11.04$\% to $1.138$\%.

## Spherical Gaussian

Answer\\

Likelihood,
\begin{gather*}
    L(\mu, \sigma^2) = \Pi_{i=1}^N\left(\frac{1}{(2\pi\sigma^2)^\frac{p}{2}}e^\frac{-1}{2\sigma^2}(x_i - \mu)^T(x_i - \mu)\right)\\
    log L = \sum_{i=1}^N\frac{-p}{2}log(2\pi\sigma^2) - \frac{1}{2\sigma^2}(x_i - \mu)^T(x_i - \mu)\\
\end{gather*}
To estimate the maximum $\mu$ and $\sigma$, we differentiate the log likelihood with respect to $\mu$ and $\sigma$ and equate to $0$ i.e.,
\begin{gather*}
    \frac{\partial log L}{\partial \mu} = 0 \\
    \implies \sum_{i=1}^N \frac{1}{\sigma^2}(x_i - \mu) = 0\\
    \sum_{i=1}^N x_i = N\mu \\
    \therefore \hat{\mu}_{MLE} = \frac{1}{N}\sum_{i=1}^N x_i\\
    \frac{\partial log L}{\partial \sigma} = 0\\
    \implies \sum_{i=1}^N\left(-\frac{p}{2\sigma^2} + \frac{1}{2\sigma^4}(x_i - \mu)^T(x_i - \mu)\right) = 0\\
    -Np\sigma^2 + \sum_{i=1}^N(x_i - \mu)^T(x_i - \mu) = 0\\
    \therefore \hat{\sigma}_{MLE} = \sqrt{\frac{1}{Np}\sum_{i=1}^N(x_i - \hat{\mu}_{MLE})^T(x_i - \hat{\mu}_{MLE})}
\end{gather*}

6- **3pt** Implement in Python a well commented function **predict** that takes as input a dataset $\mathbf{X_{\text{test}}}$ in the same dimensions as $\mathbf{X}$ and return the predictions.   Write down the mean squared error (https://en.wikipedia.org/wiki/Mean_squared_error) of your predictions. Then on the same plot with legends, show:

 a) A scatter plot of the first feature of $\mathbf{X_{\text{test}}}$ (x-axis) and the diabetes progression $\mathbf{y_{\text{test}}}$ 
 
 b) A plot of your prediction for $\mathbf{X_{\text{test}}}$

```python
# Make it possible to show plots in the notebooks.
%matplotlib inline
import numpy as np
from sklearn import datasets
import matplotlib
import matplotlib.pyplot as plt



def fit_linear_with_regularization(X, y, alpha):
    # Return w* = inv(X^TX + alphaI)X^Ty
    XTX = np.dot(X.T, X)
    I = np.eye(len(XTX))
    inv_matrix = np.linalg.inv(XTX + alpha*I)
    w_star = np.dot(np.dot(inv_matrix, X.T), y)
    return w_star     


def predict(X_test, w):
    # Return y = Xw
    return np.dot(X_test, w) 


def plot_prediction(X_test, y_test, y_pred):
    # Scatter plot the first feature of X_test (x-axis) and y_test (y-axis)
    plt.scatter(X_test[:, 1], y_test, label='y_test') 
  
    # Plot y_pred using the first feature of X_test as x-axis
    plt.scatter(X_test[:, 1], y_pred, label='y_pred')
    
    plt.legend()
  
    # Compute the mean squared error (MSE in matrix form = 1/n * eTe)
    e = y_test - y_pred 
    n = len(y_test)
    mean_squared_error = (1/n)*np.dot(e.T, e)
    return mean_squared_error
  
# Load the diabetes dataset
X, y = datasets.load_diabetes(return_X_y=True)

# Append 1 to the beginning of X
ones_column = np.ones((len(X), 1))
X = np.c_[ones_column, X]


# Split the dataset into training and test set
num_test_elements = 20

X_train = X[:-num_test_elements]
X_test = X[-num_test_elements:]

y_train = y[:-num_test_elements]
y_test = y[-num_test_elements:]


# Set alpha
alpha = 0.01

# Train using linear regression with regularization and find optimal model
w = fit_linear_with_regularization(X_train, y_train, alpha)


# Make predictions using the testing set X_test
y_pred = predict(X_test, w)



# Plots and mean squared error
error = plot_prediction(X_test, y_test, y_pred)



print('Mean Squared error is ', error)


# Show the plot
plt.show();



```

     &= -\frac{\gamma}{2}\mathbf{y}^T\mathbf{y}+\gamma\mathbf{y}^T\mathbf{Xw}-\frac{\gamma}{2}\mathbf{w}^T(\mathbf{X}^T\mathbf{X})\mathbf{w} - \frac{1}{2}\mathbf{w}^T\gamma\mathbf{Iw}
    \end{split}
\end{equation*}
